## 8. Evaluate Model Performance

### Evaluation Metrics for Text-to-Image Synthesis

```python
# Initialize loss functions
loss_fn = ResAttnGANLoss(
    loss_type='hinge',
    lambda_perceptual=1.0,
    lambda_attn=1.0,
    use_attention_loss=True
)

print("=" * 60)
print("TRAINING Res_AttnGAN")
print("=" * 60)

# Setup optimizers
lr_g = 0.0002
lr_d = 0.0002
beta1, beta2 = 0.5, 0.999

optimizer_g = optim.Adam(
    generator.parameters(),
    lr=lr_g,
    betas=(beta1, beta2)
)

optimizer_d = optim.Adam(
    discriminator.parameters(),
    lr=lr_d,
    betas=(beta1, beta2)
)

print(f"✓ Optimizers configured")
print(f"  - Generator LR: {lr_g}")
print(f"  - Discriminator LR: {lr_d}")

# Training parameters
NUM_EPOCHS = 5  # Short training for demo
SAVE_INTERVAL = 2
PRINT_INTERVAL = 10

# Training metrics
metrics = {
    'loss_g': [],
    'loss_d': [],
    'loss_perceptual': []
}

# Training loop (simplified for demo)
print(f"\nTraining for {NUM_EPOCHS} epochs...")
print(f"Batches per epoch: {len(train_loader)}")

for epoch in range(NUM_EPOCHS):
    epoch_loss_g = 0
    epoch_loss_d = 0
    
    pbar = tqdm(enumerate(train_loader), total=len(train_loader), 
                desc=f"Epoch {epoch+1}/{NUM_EPOCHS}")
    
    for batch_idx, batch in pbar:
        real_images = batch['image'].to(device)
        batch_size = real_images.size(0)
        
        # Generate random embeddings (in real scenario, encode actual captions)
        sent_embs = torch.randn(batch_size, 256).to(device)
        word_embs = torch.randn(batch_size, 20, 256).to(device)
        
        # ============= Train Discriminator =============
        noise = torch.randn(batch_size, 100).to(device)
        
        # Generate fake images
        with torch.no_grad():
            fake_images, _, attn_maps = generator(noise, sent_embs, word_embs)
        
        # Discriminate
        real_logits_list, real_text_logits_list, _ = discriminator(real_images, sent_embs)
        fake_logits_list, fake_text_logits_list, _ = discriminator(fake_images, sent_embs)
        
        # Discriminator loss (simplified)
        d_loss = 0
        for real_logits, fake_logits in zip(real_logits_list, fake_logits_list):
            d_loss += loss_fn.discriminator_loss(real_logits, fake_logits)
        
        # Update discriminator
        optimizer_d.zero_grad()\n        d_loss.backward()
        torch.nn.utils.clip_grad_norm_(discriminator.parameters(), 1.0)
        optimizer_d.step()
        
        # ============= Train Generator =============\n        noise = torch.randn(batch_size, 100).to(device)
        fake_images, _, attn_maps = generator(noise, sent_embs, word_embs)
        
        # Discriminate generated images
        fake_logits_list, fake_text_logits_list, _ = discriminator(fake_images, sent_embs)
        
        # Generator loss (simplified)
        g_loss = 0
        for fake_logits in fake_logits_list:
            g_loss += loss_fn.generator_loss(fake_logits, fake_images, real_images, attn_maps)
        
        # Update generator
        optimizer_g.zero_grad()
        g_loss.backward()
        torch.nn.utils.clip_grad_norm_(generator.parameters(), 1.0)
        optimizer_g.step()
        
        epoch_loss_g += g_loss.item()
        epoch_loss_d += d_loss.item()
        
        if batch_idx % PRINT_INTERVAL == 0:
            pbar.set_postfix({
                'loss_d': f'{d_loss.item():.4f}',
                'loss_g': f'{g_loss.item():.4f}'
            })
    
    # Average losses
    avg_loss_g = epoch_loss_g / len(train_loader)
    avg_loss_d = epoch_loss_d / len(train_loader)
    
    metrics['loss_g'].append(avg_loss_g)
    metrics['loss_d'].append(avg_loss_d)
    
    print(f"Epoch {epoch+1} - Loss D: {avg_loss_d:.4f}, Loss G: {avg_loss_g:.4f}")

print(f"\n✓ Training completed for {NUM_EPOCHS} epochs")

# Plot training metrics
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(metrics['loss_g'], label='Generator Loss', linewidth=2, marker='o')
axes[0].plot(metrics['loss_d'], label='Discriminator Loss', linewidth=2, marker='s')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Training Losses')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(metrics['loss_g'], label='Generator', linewidth=2, marker='o', color='#51CF66')
axes[1].fill_between(range(len(metrics['loss_g'])), metrics['loss_g'], alpha=0.3, color='#51CF66')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Generator Loss')
axes[1].set_title('Generator Training Progress')
axes[1].grid(True, alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 5. Design Enhanced Res_AttnGAN Model

### Solutions to AttnGAN Drawbacks:

#### 1. **Residual Connections** → Better Gradient Flow
   - Add skip connections throughout the network
   - Formula: `output = f(x) + x`
   - Benefit: Gradients flow directly through skip paths, preventing vanishing gradients

#### 2. **Hierarchical Attention** → Enhanced Text-Image Alignment
   - Multi-level attention at different feature scales
   - Word-level and phrase-level attention integration
   - Benefit: Better capture of complex relationships between text and image regions

#### 3. **Dense Feature Fusion** → Preserve Spatial Information
   - Use DenseNet-style concatenations between scales
   - Connect low-level and high-level features
   - Benefit: Retain fine-grained details through feature reuse

#### 4. **Residual Spatial Attention** → Stable Attention Weights
   - Attention computation with residual pathways
   - Improved weight normalization and regularization
   - Benefit: More stable and meaningful attention patterns

#### 5. **Spectral Normalization** → Training Stability
   - Apply spectral norm to discriminator weights
   - Lipschitz constraint on discriminator
   - Benefit: Smoother loss landscape and better gradient signals

#### 6. **Multi-scale Residual Features** → Enhanced Detail Preservation
   - Residual blocks at each generation stage
   - Feature pyramids for multi-scale processing
   - Benefit: Better preservation of spatial details and textures

## 4. Analyze AttnGAN Model and Identify Drawbacks

### Drawbacks of Original AttnGAN:

#### 1. **Gradient Flow Issues**
   - Problem: Vanishing/exploding gradients through deep networks
   - Impact: Training instability and slow convergence
   - Root Cause: No skip connections to preserve gradient pathways

#### 2. **Spatial Attention Bottleneck**
   - Problem: Single attention layer may not capture complex text-image relationships
   - Impact: Limited ability to align fine-grained details
   - Root Cause: Flat attention mechanism without hierarchical structure

#### 3. **Feature Information Loss**
   - Problem: Upsampling operations lose spatial details
   - Impact: Generated images lack fine-grained details and textures
   - Root Cause: Unidirectional feature flow without feature fusion across scales

#### 4. **Vanishing Attention Weights**
   - Problem: Softmax attention weights become too uniform or too sparse
   - Impact: Attention mechanism becomes ineffective in later stages
   - Root Cause: Numerical instability in attention weight computation

#### 5. **Limited Text-Image Correspondence**
   - Problem: Weak alignment between generated image regions and text words
   - Impact: Generated images may not accurately represent descriptions
   - Root Cause: Single-stage attention without hierarchical word-level guidance

#### 6. **Training Instability**
   - Problem: Mode collapse and training divergence
   - Impact: Low diversity in generated images
   - Root Cause: Insufficient regularization and discriminator gradient signals
